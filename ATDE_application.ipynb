```python
from ATDE import ATDE
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# 1. Qin Dynasty (221 BC - 207 BC)
- Our data starts from 361 BC

```python
qin_count = ATDE(
    video_path='Condensed Videos/Qin_Condensed.mp4',
    output_path='Validation/Qin_validation.mp4',
    main_colors=[[153, 101, 70], [193, 135, 101], [229, 207, 193]],
    start_year=-361,
    end_year=-206,
    hsv_range=10,
    lower_sv=100,
    upper_sv=255,
    min_neighbors=5
)

print(qin_count)
```

```python
years = sorted(qin_count.keys())
pixels = [qin_count[year] for year in years]

sns.set_theme(style="whitegrid", font="serif", font_scale=1.3)

plt.figure(figsize=(14, 6))
plt.plot(
    years, pixels,
    color="#0C1A18",
    linewidth=2.2
)

plt.title("Territorial Extent of the Qin Dynasty (361 BCE  206 BCE)", pad=15)
plt.xlabel("Year", labelpad=10)
plt.ylabel("Pixel Count (Territorial Extent)", labelpad=10)

plt.grid(True, linestyle="--", linewidth=0.7, alpha=0.5)
plt.xticks(np.arange(-360, -205, 25), rotation=45)
plt.xlim(-361, -206)
sns.despine()
plt.tight_layout()

plt.show()
```

# Territorial Trends Over Time

```python
dynasties = {
    'Qin': qin_count,
    'Han': han_count,
    'Jin': jin_count,
    'Sui': sui_count,
    'Tang': tang_count,
    'Song': song_count,
    'Jurchen Jin': jurchen_jin_count,
    'Yuan': yuan_count,
    'Ming': ming_count,
    'Qing': qing_count
}

# Scale factors
scale_factors = {
    'Qin': 0.113,
    'Han': 0.684,
    'Jin': 0.623,
    'Sui': 0.622,
    'Tang': 1.0,
    'Song': 0.589,
    'Jurchen Jin': 0.632,
    'Yuan': 0.698,
    'Ming': 0.628,
    'Qing': 0.684,
}

# Colors for each dynasty
colors = {
    'Qin': '#8B4513',
    'Han': '#1f77b4',
    'Jin': '#ff7f0e',
    'Sui': '#bcbd22',
    'Tang': '#2ca02c',
    'Song': '#e377c2',
    'Jurchen Jin': '#17becf',
    'Yuan': '#7f7f7f',
    'Ming': '#d62728',
    'Qing': '#8c564b'
}


palette = sns.color_palette("colorblind", n_colors=10)
colors = dict(zip(dynasties.keys(), palette))

# Compute global peak across all dynasties (after scaling)
all_peaks = [
    np.max(np.array(list(data.values())) * scale_factors[name])
    for name, data in dynasties.items() if data
]
global_peak = max(all_peaks) if all_peaks else 1

plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams.update({
    "axes.edgecolor": "black",
    "axes.linewidth": 1.2,
    "axes.labelsize": 13,
    "axes.titlesize": 15,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "legend.fontsize": 11,
    "font.family": "serif",
    "grid.alpha": 0.2
})

plt.figure(figsize=(16, 8))

for dynasty, data in dynasties.items():
    if not data:
        continue
    years = np.array(sorted(data.keys()))
    pixels = np.array([data[y] for y in years])
    scaled = pixels * scale_factors[dynasty]
    normalized = scaled / global_peak
    
    plt.plot(
        years, normalized,
        label=dynasty,
        linewidth=2.2,
        color=colors[dynasty],
        alpha=0.9
    )

plt.title('Territorial Extent of Major Chinese Dynasties', pad=20)
plt.xlabel('Year', labelpad=8)
plt.ylabel('Relative Territorial Extent (0-1)', labelpad=8)
plt.xticks(np.arange(-360, 1901, 100))
plt.xlim(-360, 1960)
plt.ylim(0, 1.05)

for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.12),
    ncol=5,
    frameon=False,
    handlelength=2.5
)

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
